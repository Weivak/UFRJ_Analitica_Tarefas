### Analise Exploratória
Perguntas:
- quais comunidades são mais afetadas por cada pergunta
- qual pais possui mais preconceito
- qual tipo de discriminação mais sofrida
- qual grupos mais sofre com cada tipo de discriminação

```python
import matplotlib.pyplot as plt
import seaborn as sns

# plotando o gráfico de barras com a quantidade de respostas por país
plt.figure(figsize=(20,10))
sns.countplot(x='pais_ingles', data=df_lgbt)
plt.title('Quantidade de respostas por país')
plt.show()

# plotando o gráfico de barras com a quantidade de respostas por grupo da comunidade
plt.figure(figsize=(20,10))
sns.countplot(x='grupo', data=df_lgbt)
plt.title('Quantidade de respostas por grupo da comunidade')
plt.show()
```

```python
# Grafico de barras com a quantidade de respostas para cada pergunta
plt.figure(figsize=(20,10))
sns.countplot(y='pergunta', data=df_lgbt)
plt.title('Quantidade de respostas para cada pergunta')
```

```python
# having our X and Y for Bar & Pie chart ready
grupos = df_lgbt['grupo'].value_counts()
lables = grupos.index
values = grupos.values

# plotting a figure
fig=plt.figure()
fig.set_figheight(10)
fig.set_figwidth(20)

a_fig=fig.add_subplot(1,2,1)
l1=plt.bar(lables, values)
l1[0].set_color('r')
plt.xticks(fontsize=10, rotation=30)
plt.xlabel('Grupos LGBTQIA+')
plt.ylabel('Total de pessoas')
plt.title('Metadados LGBTQIA+',fontsize=15)
```
