### Limpeza dos Dados

```python
# unindo os dataframes
import pandas as pd
import pandas_profiling

# aqui ocorre a união dos dataframes relacionados ao tema LGBT, exceto o df_especifico_transgenero
# pois esse dataframe contem dados especificos sobre a população transgênero e será
# analisado em separadamente
df_lgbt = pd.concat([df_consciencia_direitos,df_cotidiano,df_discriminacao,df_violencia_abuso], axis=0)
```

### Analise Exploratória


```python
import matplotlib.pyplot as plt
import seaborn as sns

# plotando o gráfico de barras com a quantidade de respostas por país
plt.figure(figsize=(20,10))
sns.countplot(x='pais_ingles', data=df_lgbt)
plt.title('Quantidade de respostas por país')
plt.show()

# plotando o gráfico de barras com a quantidade de respostas por grupo da comunidade
plt.figure(figsize=(20,10))
sns.countplot(x='grupo', data=df_lgbt)
plt.title('Quantidade de respostas por grupo da comunidade')
plt.show()
```

```python
grupos = df_lgbt['grupo'].value_counts()
lables = grupos.index
values = grupos.values

fig=plt.figure()
fig.set_figheight(5)
fig.set_figwidth(10)

a_fig=fig.add_subplot(1,2,1)
l1=plt.bar(lables, values)
l1[0].set_color('r')
plt.xticks(fontsize=10, rotation=30)
plt.xlabel('Grupos LGBTQIA+')
plt.ylabel('Total de pessoas')
plt.title('Metadados LGBTQIA+',fontsize=15)

```

```python
# respostas por grupos para a pergunta 1
pergunta_1 = "In the country where you live, is there a law that forbids discrimination against persons because of their sexual orientation when applying for a job?"
pergunta = df_lgbt.loc[df_lgbt['pergunta'] == pergunta_1]
respostas = pergunta['resposta'].value_counts()

lables = respostas.index
values = respostas.values

colors = ['r', 'b', 'g', 'y']

fig, ax = plt.subplots(figsize = (10,6))
ax.grid(False)
plt.style.use('ggplot')
  
plt.pie(values,explode= (0.1,0,0) , labels = lables, colors=colors, 
        startangle=90, shadow = False,
        radius = 1.2, autopct = '%1.1f%%')
plt.legend(title="Answer",
          loc="center left",
          bbox_to_anchor=(1.5, 0, 0.8, 1))
plt.title(pergunta_1,fontsize=15)
plt.show()
```

```python
# respostas por grupos para a pergunta 2
pergunta_2 = "In your opinion, how widespread is offensive language about lesbian, gay, bisexual and/or transgender people by politicians in the country where you live?"
pergunta = df_lgbt.loc[df_lgbt['pergunta'] == pergunta_1]
respostas = pergunta['resposta'].value_counts()

lables = respostas.index
values = respostas.values

colors = ['r', 'b', 'g', 'y','c']

fig, ax = plt.subplots(figsize = (10,6))
ax.grid(False)
plt.style.use('ggplot')
  
plt.pie(values,explode= (0,0,0,0,0) , labels = lables, colors=colors, 
        startangle=90, shadow = False,
        radius = 1.2, autopct = '%1.1f%%')
plt.legend(title="Answer",
          loc="center left",
          bbox_to_anchor=(1.5, 0, 0.8, 1))
plt.title(pergunta_2,fontsize=15)
plt.show()
```

```python
# respostas por grupos para a pergunta 3
pergunta_3 = "Have you ever experienced any of the following situations when using or trying to access healthcare services as a L, G, B or T person?"
pergunta = df_lgbt.loc[df_lgbt['pergunta'] == pergunta_3]
respostas = pergunta['resposta'].value_counts()

lables = respostas.index
values = respostas.values

colors = ['r', 'b', 'g', 'y','c','m','k']

fig, ax = plt.subplots(figsize = (10,6))
ax.grid(False)
plt.style.use('ggplot')
  
plt.pie(values,explode= (0,0,0,0,0,0,0,0,0) , labels = lables, colors=colors, 
        startangle=90, shadow = False,
        radius = 1, autopct = '%1.1f%%')
plt.legend(title="Answer",
          loc="center left",
          bbox_to_anchor=(1.5, 0, 0.8, 1))
plt.title(pergunta_3,fontsize=15)
plt.show()
```
